```python
import numpy as np
import matplotlib.pyplot as plt
# Set the random seed based on the last two digits of your roll number
np.random.seed(36)  # Change 36 to your last two digits
# Generate random coefficients for a 5th degree polynomial
coefficients = np.random.randint(-10, 10, 6)  # 6 coefficients for a 5th degree polynomial
# Define the polynomial function using numpy's poly1d
polynomial = np.poly1d(coefficients)
# Generate x values for plotting
x = np.linspace(-5, 5, 1000)
# Compute y values using the polynomial function
y = polynomial(x)
# Plot the polynomial
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Polynomial Curve')
# Find the maxima and minima of the polynomial
maxima = x[np.argmax(y)]
minima = x[np.argmin(y)]
# Annotate the plot to label the maxima and minima points
plt.scatter([maxima, minima], [polynomial(maxima), polynomial(minima)], color='red')
plt.annotate(f'Maxima: ({maxima:.2f}, {polynomial(maxima):.2f})', (maxima, polynomial(maxima)),
             xytext=(-30, 30), textcoords='offset points', arrowprops=dict(facecolor='black', arrowstyle='->'))
plt.annotate(f'Minima: ({minima:.2f}, {polynomial(minima):.2f})', (minima, polynomial(minima)),
             xytext=(-30, -30), textcoords='offset points', arrowprops=dict(facecolor='black', arrowstyle='->'))
# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of a 5th Degree Polynomial with Maxima and Minima Annotation')
plt.legend()
plt.grid(True)
plt.show()
```
