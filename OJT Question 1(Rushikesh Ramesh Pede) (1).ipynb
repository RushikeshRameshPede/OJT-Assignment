```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
# Load the dataset
iris_df = pd.read_csv(r"C:\Users\User\Downloads\iris.csv")
# Display the first few rows of the dataset
print(iris_df.head())
# Summary statistics
print("\nSummary Statistics:")
print(iris_df.describe())
# Data Visualization
# Histograms for numerical columns
iris_df.hist(figsize=(10, 8))
plt.tight_layout()
plt.show()
# Box plots for numerical columns
plt.figure(figsize=(10, 8))
sns.boxplot(data=iris_df)
plt.title("Box plots for Numerical Columns")
plt.show()
# Scatter plots for numerical columns against each other
sns.pairplot(iris_df, kind="scatter", diag_kind="kde")
plt.show()
# Load the dataset
iris_df = pd.read_csv(r"C:\Users\User\Downloads\iris.csv")
# Extract numerical columns
numerical_columns = iris_df.select_dtypes(include=['float64', 'int64'])
# Calculate mean for each numerical column
mean_values = numerical_columns.mean()
# Calculate median for each numerical column
median_values = numerical_columns.median()
# Calculate mode for each numerical column
mode_values = numerical_columns.mode().iloc[0]
# Calculate standard deviation for each numerical column
std_dev_values = numerical_columns.std()
# Print the results
print("Mean:")
print(mean_values)
print("\nMedian:")
print(median_values)
print("\nMode:")
print(mode_values)
print("\nStandard Deviation:")
print(std_dev_values)
```
